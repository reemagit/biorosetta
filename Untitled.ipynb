```python
import requests
from tqdm import tqdm


def download(url: str, fname: str):
    resp = requests.get(url, stream=True)
    total = int(resp.headers.get('content-length', 0))
    # Can also replace 'file' with a io.BytesIO object
    with open(fname, 'wb') as file, tqdm(
        desc=fname,
        total=total,
        unit='iB',
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
        for data in resp.iter_content(chunk_size=1024):
            size = file.write(data)
            bar.update(size)
```

```python
data = add
id_orig = 'id1'
id_synonym = 'id2'
id_out = 'id3'
pruned = data[data[id_orig].notna() & data[id_synonym].notna()]
pruned = pruned[~pruned.duplicated(subset=[id_orig, id_synonym])]
pruned = pruned[~pruned.duplicated(subset=[id_synonym], keep=False)] # we just remove inconsistent synonyms
pruned_curr = pruned.loc[~pruned[id_synonym].isin(lookup.index)].copy()
pruned_curr[id_out] = lookup.loc[pruned[id_orig].values].values
pruned_curr = pruned_curr.set_index(id_synonym)[id_out].squeeze()
new_lookup = pd.concat([lookup,pruned_curr])
```
